In this lab, we will work with the [SF Salaries ](https://www.kaggle.com/kaggle/sf-salaries) database from Kaggle.

**Q1** : Import the pandas library and read the Salaries.csv file.
Look at the header of the DataFrame, observe the data, and the explanatory variables.

```python
import pandas as pd

df = pd.read_csv("/content/Salaries.csv")
df.head()
```

**Q2** : Use the info() method to see how many entries you have.




```python
df.info()
```

**Q3** : What is the average of BasePay?

```python
pd.to_numeric(df["BasePay"], errors="coerce").mean()
```

**Q4** :What is the maximum value of OvertimePay in this dataset?

```python
pd.to_numeric(df["OvertimePay"], errors="coerce").max()
```

**Q5** :What is the JobTitle of JOSEPH DRISCOLL (case sensitive)?

```python
df[df["EmployeeName"] == "JOSEPH DRISCOLL"]["JobTitle"]
```

**Q6** :What is the TotalPayBenefits of JOSEPH DRISCOLL?

```python
df[df["EmployeeName"] == "JOSEPH DRISCOLL"]["TotalPayBenefits"]
```

**Q7** : What is the name of the person who is paid the most? Display all of their details.



```python
df[df["TotalPayBenefits"] == df["TotalPayBenefits"].max()]
```

**Q8** : What is the name of the person who is paid the least? Display all of their details.


```python
df[df["TotalPayBenefits"] == df["TotalPayBenefits"].min()]
```

**Q9** : What is the average BasePay per year?

```python
pd.to_numeric(df["BasePay"], errors="coerce").groupby(df["Year"]).mean()
```
